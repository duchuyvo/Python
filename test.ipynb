```python
import pandas as pd

data = pd.read_csv("Customers.csv")
```

```python
#hiển thị số lượng dòng và cột của DataFrame hoặc Series, tên và kiểu dữ liệu của các cột, số lượng giá trị khác null
data.info()
#Phương thức này được sử dụng để tính toán một số thống kê mô tả bao gồm số lượng, trung bình, độ lệch chuẩn
data.describe()
#hiển thị 10 hàng đầu tiên trong DataFrame.
data.head()
```

```python
#sử dụng để lấy danh sách các giá trị duy nhất (unique) trong cột "Gender" của đối tượng DataFrame
unique_values_gender = data['Gender'].unique()
print(unique_values_gender)
```

```python
from sklearn.preprocessing import LabelEncoder
# sử dụng để chuyển đổi các giá trị chuỗi (string) của cột "Gender"
le = LabelEncoder()
data['Gender'] = le.fit_transform(data['Gender'])
# hiển thị 10 dòng đầu tiên
data.head(10)
```

```python
#được sử dụng để điền giá trị trống (missing value) trong cột "Profession" thay bằng bằng giá trị chuỗi "No Profession"
data['Profession'].fillna('No Profession', inplace=True)
#trích xuất danh sách các giá trị duy nhất (unique values) trong cột "Profession" của đối tượng DataFrame "data".
unique_values_profession = data['Profession'].unique()
print(unique_values_profession)
```

```python
from sklearn.preprocessing import LabelEncoder
# tạo một đối tượng LabelEncoder mới được gán cho biến le.
le = LabelEncoder()
# sử dụng fit_transform() của LabelEncoder le để chuyển đổi giá trị chuỗi trong cột "Profession" thành các số nguyên. 
#Kết quả được gán lại cho cột "Profession"
data['Profession'] = le.fit_transform(data['Profession'])
data.head(10)
```

```python
#trích xuất các giá trị duy nhất trong cột "Profession" gán cho unique_values_profession.
unique_values_profession = data['Profession'].unique()
print(unique_values_profession)
#lưu đối tượng DataFrame "data" sau khi tiền xử lý dữ liệu vào tệp "Customers_encode.csv".
data.to_csv("Customers_encode.csv", index=False)
```

```python
#đọc đối tượng DataFrame đã được tiền xử lý dữ liệu từ tệp "Customers_encode.csv" và gán nó cho biến data_encode.
data_encode = pd.read_csv("Customers_encode.csv")
#info() để hiển thị thông tin về các cột và kiểu dữ liệu 
data_encode.info()
#describe() để hiển thị các thống kê mô tả về dữ liệu.
data_encode.describe()
#head() hiển thị 5 dòng đầu tiên của dữ liệu.
data_encode.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error, r2_score, confusion_matrix
import matplotlib.pyplot as plt
import numpy as np

datas = pd.read_csv("Customers_encode.csv")
# drop() để loại bỏ cột "Work Experience" 
# Trích xuất đặc trưng
X = datas.drop("Work Experience", axis=1)
y = datas["Work Experience"]

# Tách dữ liệu
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

<h1 style="color: black">Linear Regression Model</h1>

```python
# Mô hình hồi quy tuyến tính
linearReg = LinearRegression()
linearReg.fit(X_train, y_train)

# Đưa ra dự đoán về dữ liệu thử nghiệm của Mô hình hồi quy tuyến tính
y_pred_linearReg = linearReg.predict(X_test)
print(y_pred_linearReg)
print()

# Độ chính xác của dữ liệu đối với Mô hình hồi quy tuyến tính
print('Mean squared error: ', mean_squared_error(y_test, y_pred_linearReg))
print()
print('R2 score', r2_score(y_test, y_pred_linearReg))
print()

# Confusion matrix to Linear Regression Model
y_pred_binary_linearReg = np.where(y_pred_linearReg > 0.5, 1, 0)
confusion_matrix_linearReg = confusion_matrix(y_test, y_pred_binary_linearReg)
print("Confusion matrix to Linear Regression Model: ")
print(confusion_matrix_linearReg)
```

# Logistic Regression Model

```python
# Mô hình hồi quy logictic (Logistic Regression) 
logisticReg = LogisticRegression()
logisticReg.fit(X_train, y_train)

# Đưa ra dự đoán về dữ liệu thử nghiệm của Mô hình hồi quy logistic
y_pred_logisticReg = logisticReg.predict(X_test)
print(y_pred_logisticReg)
print()

# Độ chính xác của dữ liệu đối với Mô hình hồi quy logistic
print("Accuracy to Logistic Regression Model: ", accuracy_score(y_test, y_pred_logisticReg))
print()

# Thống kê kết quả của mô hình Logistic Regression
confusion_matrix_logisticReg = confusion_matrix(y_test, y_pred_logisticReg)
print("Confusion matrix to Logistic Regression Model: ")
print(confusion_matrix_logisticReg)
```

# Decision Tree Model

```python
# Mô hình cây quyết định
decisionTree = DecisionTreeClassifier()
decisionTree.fit(X_train, y_train)

# Đưa ra dự đoán về dữ liệu thử nghiệm của Mô hình cây quyết định
y_pred_decisionTree = decisionTree.predict(X_test)
print(y_pred_decisionTree)
print()

# Độ chính xác của dữ liệu đối với Mô hình cây quyết định
print("Accuracy to Decision Tree Model: ", accuracy_score(y_test, y_pred_decisionTree))
print()

# Thống kê kết quả Decision Tree Model
confusion_matrix_decisionTree = confusion_matrix(y_test, y_pred_decisionTree)
print("Confusion matrix to Decision Tree Model: ")
print(confusion_matrix_decisionTree)
#in cây quyết định
```

# K-Nearest Neighbors Model

```python
# Mô hình K-Nearest Neighbors 
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

# Đưa ra dự đoán về dữ liệu thử nghiệm của K-Nearest Neighbors Model
y_pred_knn = knn.predict(X_test)
print(y_pred_knn)
print()

# Đo độ chính xác của mô hình K-Nearest Neighbors Model
print("Accuracy to K-Nearest Neighbors Model: ", accuracy_score(y_test, y_pred_knn))
print()

# Thống kê kết quả của mô hình K-Nearest Neighbors
confusion_matrix_knn = confusion_matrix(y_test, y_pred_knn)
print("Confusion matrix to K-Nearest Neighbors Model: ")
print(confusion_matrix_knn)
```
